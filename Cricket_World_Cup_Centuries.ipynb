```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Read** **Dataset**

```python
df= pd.read_csv('/content/All-Cricket-World-Cup-Centuries.csv')
df
```

# **Array** **of** **Team**

```python
arr=np.array(df['Team'])
arr
```

# **Print** **Unique** **Values**

```python
x=np.array(df['Team'].unique())
y=np.array(df['Team'].value_counts(sort=False))
print(x)
print(y)
```

# **Bar plot of count of unique teams**

```python
x=np.array(df['Team'].unique())
y=np.array(df['Team'].value_counts(sort=False))
fig, ax = plt.subplots(figsize=(20, 10))
sns.barplot(data=df,x=x,y=y,ax=ax)
```

# **Counting loss of Team**

```python
df['Result_Category'] = df['Result'].apply(lambda x: 'Win' if x == 'won' else 'Loss')
result_counts = df.groupby(['Team', 'Result_Category']).size().unstack(fill_value=0)
print(result_counts)
```

# **Plot of win loss and tie**

```python
team_results=df.groupby(['Team','Result']).size().unstack().fillna(0)
team_results=team_results[['Won','Lost','Tied']].fillna(0)
team_results.plot(kind='bar',stacked=False,color=['g','r','b'],figsize=(20,8))
print(team_results)
```

# **Violin Plot**
